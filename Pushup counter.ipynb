```python
import cvzone
import cv2
import mediapipe as mp
from cvzone.PoseModule import PoseDetector
import mediapipe as mp
import numpy as np
import math


# creating a veriable for pushup count:
Direction= 0
Pushup_count = 0
```

```python
cap=cv2.VideoCapture("Pushup_2.mp4")

Pose_detector= PoseDetector(detectionCon=0.8,trackCon=0.8)

while cap.isOpened():
    _, img = cap.read()
    img=cv2.resize(img,(512,640))
    
    frame=Pose_detector.findPose(img)
    lmList, bbox = Pose_detector.findPosition(img,draw=False)    
    
    if len(lmList):
        
        
        # for right hands:
        A1=Pose_detector.findAngle(img,12,14,16)
        # for left hands:
        A2=Pose_detector.findAngle(img,15,13,11)
        
        # percentage calculater:
        per_val1 = round(np.interp(A1,(54,179),(100,0)))
        per_val2 = round(np.interp(A2,(49,179),(100,0)))
        
        #bar value for right side:
        bar_val1=round(np.interp(per_val1,(0,100),(75+410,75)))
        
        #bar value for left side:
        bar_val2=round(np.interp(per_val2,(0,100),(140+350,140)))
        
        #creating a Rectengle right side:
        cv2.rectangle(img,(440,bar_val1),(440+25,75+410),(0,255,0),cv2.FILLED)
        cv2.rectangle(img,(440,75),(440+25,75+410),(0,0,0),2)
        
        
        #creating a Rectengle left side:
        cv2.rectangle(img,(20,bar_val2),(20+25,140+349),(0,255,0),cv2.FILLED)
        cv2.rectangle(img,(20,140),(20+25,140+349),(0,0,0),2)
        
        # counting pushup:
        
        if Direction == 0:
            
            if per_val1 == 100 and per_val2 == 100:
                Pushup_count += 1
                Direction = 1
                colour = (0,255,0)
            
            
        elif Direction == 1:
            if per_val1 == 0 and per_val2 == 0:
                Pushup_count += 0
                Direction = 0
                colour = (0,255,0)
        else:
            colour = (0,0,255)
        
        cvzone.putTextRect(img,f'Pushup_count: {Pushup_count}',(130,40),1.5,2,colorT=(255, 255, 255),colorR=(255, 0, 255),border=2,colorB=())
        
    cv2.imshow("Pushup",img)
    
    
    
    if cv2.waitKey(1)== ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```
